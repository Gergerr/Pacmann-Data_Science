# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py
load_time_1 = 0
exit_time_1 = 0

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 0
exit_time_3 = 0
```

**Output - 2**

```py
Session 1 Duration (in minutes): 0
Session 2 Duration (in minutes): 0
Session 3 Duration (in minutes): 0
Average Session Duration (in minutes): 0
```

**Input - 3**

```py
load_time_1 = 3
exit_time_1 = 7

load_time_2 = -2
exit_time_2 = 5

load_time_3 = 3
exit_time_3 = 156
```

**Output - 3**

```py
Not a Valid Input
```

````python
# List Temp: Celsius, Fahrenheit, Kelvin
temp = 27
from_unit = 'celCiUs'
to_unit = 'KeLVIN'

if from_unit.lower() == 'celcius':
    if to_unit.lower() == 'kelvin':
        converted_temp= temp+273.15
        print('{} Celsius is equal to {} Kelvin'.format(temp, converted_temp))
    elif to_unit.lower() == 'fahrenheit':
        converted_temp= (temp * 1.8) + 32
        print('{} Celsius is equal to {} Fahrenheit'.format(temp, converted_temp))
    else:
        print('Not a valid target')

elif from_unit.lower() == 'fahrenheit':
    if to_unit.lower() == 'kelvin':
        converted_temp= (temp - 32)/1.8 + 273.15
        print('{} Fahrenheit is equal to {} Kelvin'.format(temp, converted_temp))
    elif to_unit.lower() == 'celcius':
        converted_temp= (temp-32)/1.8
        print('{} Fahrenheit is equal to {} Celcius'.format(temp, converted_temp))
    else:
        print('Not a valid target')

elif from_unit.lower() == 'kelvin':
    if to_unit.lower() == 'fahrenheit':
        converted_temp= 1.8 * (temp - 273.15) + 32
        print('{} Kelvin is equal to {} Fahrenheit'.format(temp, converted_temp))
    elif to_unit.lower() == 'celcius':
        converted_temp= temp - 273.15
        print('{} Kelvin is equal to {} Celcius'.format(temp, converted_temp))
    else:
        print('Not a valid target')

else:
    print('Not a valid input')
````

````python
# Write your code here
nik_1 = 1571011709860003
nik_2 = 2404075109910009



nik_1= str(nik_1) # Convert ke string agar mudah di slice
nik_2= str(nik_2)

if len(nik_1) == 16 and len(nik_2) == 16:
    nik_1_date= int(nik_1[6:8])
    nik_2_date= int(nik_2[6:8])


    # GENDER
    # NIK 1
    if nik_1_date >= 1 and nik_1_date <= 31:
        nik_1_gender= 'Male'
    else:
        nik_1_gender= 'Female'
        # Kurang revisi
        nik_1_date-= 40

    # NIK 2
    if nik_2_date >= 1 and nik_2_date <= 31:
        nik_2_gender= 'Male'
    else:
        nik_2_gender= 'Female'
        # Kurang revisi
        nik_2_date-=40

    # MONTH
    month= {
        '01': 'January',
        '02': 'February',
        '03': 'March',
        '04': 'April',
        '05': 'May',
        '06': 'June',
        '07': 'July',
        '08': 'August',
        '09': 'September',
        '10': 'October',
        '11': 'November',
        '12': 'December'
    }
    month_1_num= nik_1[8:10]
    month_2_num= nik_2[8:10]
    if 0 < int(month_1_num) <= 12 and 0 < int(month_2_num) <=12:
        nik_1_month= month.get(month_1_num)
        nik_2_month= month.get(month_2_num)
    else:
        print('Invalid NIK! Check NIK again')

    # YEAR
    nik_1_year= int(nik_1[10:12])
    nik_2_year= int(nik_2[10:12])

    # Output:
    print('Date of Birth')
    print('- NIK 1: {}-{}-{}'.format(nik_1_date, month_1_num, nik_1_year))
    print('- NIK 2: {}-{}-{}'.format(nik_2_date, month_2_num, nik_2_year))
    print('\n\nGender')
    print('- NIK 1: {}'.format(nik_1_gender))
    print('- NIK 2: {}'.format(nik_2_gender))

    print('\n\nDo they have same gender?')
    if nik_1_gender.lower() == nik_2_gender.lower():
        print('Yes')
    else:
        print('No')
    print('\n\nDo they have similar month of birth?')
    if nik_1_month == nik_2_month:
        print('Yes, it is on {}'.format(nik_1_month))
    else:
        print('No, NIK 1 born on {}, while NIK 2 born on {}'.format(nik_1_month, nik_2_month))
else:
    print('Invalid NIK format')
    if len(nik_1) != 16 and len(nik_2) == 16:
        print('Length of NIK 1 is {} instead of 16, while NIK 2 is already in correct format'.format(len(nik_1)))
    elif len(nik_1) == 16 and len(nik_2) != 16:
        print('Length of NIK 1 is already in correct format, however length of NIK 2 is {} instead of 16'.format(len(nik_2)))
    else:
        print('Both NIK is not in the correct format. Length of NIK 1 is {} instead of 16, and length of NIK 2 is {} instead of 16'.format(len(nik_1), len(nik_2)))
````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

- Your task is calculate the income tax given
  - Monthly income
  - Marital status
  - The number of dependents (jumlah tanggungan)

**Input - 1**

````python
monthly_income= 4750000
marital_status= 'single'
n_dependents= 0

yearly_income= monthly_income * 12

initial_ptkp= 54000000
additional_ptkp= 0
dependent_dict= {
    0: 0,
    1: 4500000,
    2: 9000000,
    3: 13500000
}
if n_dependents in dependent_dict:
    additional_ptkp += dependent_dict[n_dependents]
else:
    additional_ptkp += dependent_dict[3]

if marital_status.lower() == 'married':
    additional_ptkp += 4500000
    m_status= 'K'
else:
    additional_ptkp += 0
    m_status= 'TK'

final_ptkp = initial_ptkp + additional_ptkp

# Penghasilan Kena Pajak variable will be named as pkp
pkp= yearly_income - final_ptkp
tingkat_ptkp= 0
if yearly_income > 54_000_000:
    if yearly_income < 60_000_000:
        pph= 0.05 * pkp
    elif yearly_income < 250_000_000:
        pph= 0.15 * pkp
    elif yearly_income < 500_000_000:
        pph= 0.25 * pkp
    elif yearly_income < 5_000_000_000:
        pph= 0.3 * pkp
    else:
        pph= 0.35 * pkp
else:
    pph= 0
print('Penghasilan neto setahun:\n{}'.format(yearly_income))
print('\nPTKP {}/{}:\n-{}'.format(m_status, n_dependents, final_ptkp))
print('\nPenghasilan Kena Pajak: \n{}'.format(pkp))
print('\nPPh 21 terutang setahun:\n{}'.format(pph))
````
